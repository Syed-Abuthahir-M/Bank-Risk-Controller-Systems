```python
days_col = ["DAYS_BIRTH", "DAYS_EMPLOYED", "DAYS_REGISTRATION", "DAYS_ID_PUBLISH", "DAYS_LAST_PHONE_CHANGE", "DAYS_DECISION"]
```

```python
# Convert negative days to positive values for all "DAYS_" columns
for col in days_col:
    data[col] = data[col].abs()
```

```python
data.head()
```

```python
# Convert days to years for better interpretation
for col in days_col:
    data[f"{col}_YEARS"] = (data[col] / 365.25).round(1)
```
