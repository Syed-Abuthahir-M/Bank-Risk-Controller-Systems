```python
corr_matrix = data[['AMT_INCOME_TOTAL', 'AMT_CREDIT_x', 'AMT_ANNUITY_x', 'AMT_GOODS_PRICE_x', 'OBS_30_CNT_SOCIAL_CIRCLE', 
                    'DEF_30_CNT_SOCIAL_CIRCLE', 'OBS_60_CNT_SOCIAL_CIRCLE', 'DEF_60_CNT_SOCIAL_CIRCLE']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='TARGET', data=data, palette='viridis')
plt.title('Target Variable Distribution')
plt.show()
```

```python
avg_credit = data.groupby('NAME_CONTRACT_TYPE_x')['AMT_CREDIT_x'].mean().reset_index()
sns.barplot(x='NAME_CONTRACT_TYPE_x', y='AMT_CREDIT_x', data=avg_credit, palette='coolwarm')
plt.title('Average Credit Amount by Contract Type')
plt.show()
```

```python
sns.countplot(y='NAME_INCOME_TYPE', data=data, order=data['NAME_INCOME_TYPE'].value_counts().index, palette='mako')
plt.title('Distribution of Income Types')
plt.show()
```

```python
sns.boxplot(data=data, x='AMT_CREDIT_x')
plt.title('Credit Amount Distribution')
plt.show()
```

```python
sns.barplot(data=data, x='CODE_GENDER', y='TARGET')
plt.title('Default Rate by Gender')
plt.show()
```

```python
sns.boxplot(data=data, x='TARGET', y='AMT_CREDIT_x')
plt.title('Credit Amount by Target')
plt.show()
```

```python
sns.scatterplot(x='AMT_INCOME_TOTAL', y='AMT_CREDIT_x', hue='TARGET', data=data, alpha=0.7)
plt.title('Income vs Credit Amount (Colored by Target)')
plt.show()
```

```python
sns.scatterplot(data=data, x='OBS_30_CNT_SOCIAL_CIRCLE', y='DEF_30_CNT_SOCIAL_CIRCLE')
plt.title('Social Circle Observation vs Default Count')
plt.show()
```

```python
sns.histplot(data['DAYS_BIRTH_YEARS'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(y='OCCUPATION_TYPE', data=data, order=data['OCCUPATION_TYPE'].value_counts().index, palette='crest')
plt.title('Distribution of Occupation Types')
plt.show()
```

```python
sns.countplot(data=data, x='FLAG_LAST_APPL_PER_CONTRACT', hue='NFLAG_INSURED_ON_APPROVAL')
plt.title('Flag Analysis')
plt.show()
```
