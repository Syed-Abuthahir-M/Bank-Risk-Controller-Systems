```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("F:\GUVI\Project\FINAL PROJECT\Final_data.csv")
```

```python
data.head()
```

```python
data['DAYS_BIRTH'] = data['DAYS_BIRTH'].abs()
data['DAYS_EMPLOYED'] = data['DAYS_EMPLOYED'].abs()
```

```python
data.head()
```

```python
data['AGE_YEARS'] = (data['DAYS_BIRTH']) / 365
data['AGE_YEARS'] = data['AGE_YEARS'].astype(int)
```

```python
data.head()
```

```python
data.drop("DAYS_BIRTH", axis=1, inplace=True)
```

```python
data.shape
```

```python
data.duplicated().sum()
```

```python
data.info()
```

```python
sns.countplot(data=data, x='TARGET')
plt.title('Target Variable Distribution')
plt.show()
```

```python
sns.countplot(data=data, x='NAME_CONTRACT_TYPE_x', palette='viridis')
plt.title('Contract Type Distribution')
plt.show()
```

```python
sns.histplot(data=data, x='AMT_INCOME_TOTAL', kde=True, bins=30)
plt.title('Income Distribution')
plt.show()
```

```python
sns.boxplot(data=data, x='AMT_CREDIT_x')
plt.title('Credit Amount Distribution')
plt.show()
```

```python
sns.barplot(data=data, x='CODE_GENDER', y='TARGET')
plt.title('Default Rate by Gender')
plt.show()
```

```python
sns.boxplot(data=data, x='TARGET', y='AMT_CREDIT_x')
plt.title('Credit Amount by Target')
plt.show()
```

```python
numerical_cols = ['AMT_INCOME_TOTAL', 'AMT_CREDIT_x', 'DAYS_BIRTH', 'DAYS_EMPLOYED']
sns.pairplot(data[numerical_cols])
plt.show()
```

```python
sns.scatterplot(data=data, x='OBS_30_CNT_SOCIAL_CIRCLE', y='DEF_30_CNT_SOCIAL_CIRCLE')
plt.title('Social Circle Observation vs Default Count')
plt.show()
```

```python
bureau_cols = ['AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY', 'AMT_REQ_CREDIT_BUREAU_WEEK',
               'AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_YEAR']
data[bureau_cols].boxplot(figsize=(10, 6))
plt.title('Credit Bureau Requests')
plt.show()
```

```python
sns.countplot(data=data, x='FLAG_LAST_APPL_PER_CONTRACT', hue='NFLAG_INSURED_ON_APPROVAL')
plt.title('Flag Analysis')
plt.show()
```
